#### Q2: What is a scatter plot? Use the following code to generate data for x and y.



A scatter plot is a sort of graph that is used to show the relationship between two variables. It is a graph that plots the values of two variables as points on a two-dimensional plane, with one variable on the x-axis and the other on the y-axis. Scatter plots are useful for discovering outliers and odd observations in data as well as recognising patterns and trends.



```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("X-AXIS")
plt.ylabel("Y_AXIS")
plt.title("GRAPH BETWEEN X AND Y")
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.



Matplotlib's subplot() method is used to construct numerous plots in the same figure. It may be used to compare and analyse multiple datasets or to exhibit several viewpoints of the same data in a single picture.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])


x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])


x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')


fig.suptitle(' Line Plots')

plt.show()
```

#### Q4: What is a bar plot? Why is it used?



A bar plot is a sort of graph that is used to present and compare the values of several categories or groupings. It is made up of rectangular bars whose lengths correspond to the values they represent. Bar charts are commonly used to compare the values of different categories or groups when displaying categorical data.

A horizontal bar plot is one in which the bars are positioned horizontally rather than vertically. It is frequently used when the category labels are lengthy or when there are several categories to display.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

#### Q5: What is a box plot? Why is it used?



A box plot is a form of graph that is used to show the distribution of a dataset. It is made up of a box and whiskers, with the box representing the interquartile range (IQR) and the whiskers representing the data range.

Box plots are used to display a dataset's central tendency, dispersion, and skewness. They can also be used to identify probable data outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)
plt.figure(figsize=(6,3))
```

```python
plt.boxplot(box2)
plt.figure(figsize=(6,3))
```
